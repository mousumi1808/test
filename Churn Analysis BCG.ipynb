```python
# Convert price_date to datetime
price_df["price_date"] = pd.to_datetime(price_df["price_date"])
```

```python
price_avg = price_df.groupby("id").agg({
    "price_off_peak_var": "mean",
    "price_peak_var": "mean",
    "price_mid_peak_var": "mean",
    "price_off_peak_fix": "mean",
    "price_peak_fix": "mean",
    "price_mid_peak_fix": "mean"
}).reset_index()
```

```python
# Merge datasets
merged_df = client_df.merge(price_avg, on="id", how="left")
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=merged_df, x="channel_sales", hue="churn",
              order=merged_df["channel_sales"].value_counts().index)
plt.title("Churn Count by Sales Channel")
plt.xlabel("Sales Channel")
plt.ylabel("Customer Count")
plt.xticks(rotation=45)
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
# 2. Boxplot: 12-month Consumption vs Churn
plt.figure(figsize=(8, 5))
sns.boxplot(data=merged_df, x="churn", y="cons_12m")
plt.title("12-Month Consumption vs Churn")
plt.xlabel("Churn")
plt.ylabel("Consumption (kWh)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load data
client_df = pd.read_csv("client_data.csv")
price_df = pd.read_csv("price_data.csv")
price_df["price_date"] = pd.to_datetime(price_df["price_date"])

# STEP 1: Calculate Monthly Electricity Consumption Proxy
client_df["monthly_consumption"] = client_df["cons_12m"] / 12

# STEP 2: Calculate Monthly % Change in Price (for each customer)
price_df.sort_values(by=["id", "price_date"], inplace=True)

# Calculate price index as average of variable + fixed (off_peak)
price_df["price_total"] = price_df[[
    "price_off_peak_var", "price_peak_var", "price_mid_peak_var",
    "price_off_peak_fix", "price_peak_fix", "price_mid_peak_fix"
]].mean(axis=1)

# Calculate % change in price month over month for each customer
price_df["pct_price_change"] = price_df.groupby("id")["price_total"].pct_change()

# STEP 3: Merge monthly price changes back to client data
# We'll take average % price change per customer (to compare with churn)
avg_price_change = price_df.groupby("id")["pct_price_change"].mean().reset_index()
avg_price_change.rename(columns={"pct_price_change": "avg_pct_price_change"}, inplace=True)

# Merge with client data
elasticity_df = client_df.merge(avg_price_change, on="id", how="left")

# STEP 4: Estimate Elasticity Proxy
# % change in quantity = monthly_consumption / average of all customers
avg_consumption = client_df["monthly_consumption"].mean()
elasticity_df["pct_qty_change"] = (elasticity_df["monthly_consumption"] - avg_consumption) / avg_consumption

# Elasticity = % change in quantity / % change in price
elasticity_df["price_elasticity"] = elasticity_df["pct_qty_change"] / elasticity_df["avg_pct_price_change"]

# Clean up infinite and NaN values
elasticity_df["price_elasticity"] = elasticity_df["price_elasticity"].replace([np.inf, -np.inf], np.nan)

# STEP 5: Tag Customers as Elastic / Inelastic
elasticity_df["elasticity_type"] = np.where(abs(elasticity_df["price_elasticity"]) > 1, "Elastic", "Inelastic")

# STEP 6: Compare Elastic vs Inelastic Churn Rates
churn_rate_by_elasticity = elasticity_df.groupby("elasticity_type")["churn"].mean().reset_index()

churn_rate_by_elasticity["churn_percent"] = churn_rate_by_elasticity["churn"] * 100
print(churn_rate_by_elasticity)
```
